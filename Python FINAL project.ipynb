```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset_1 = pd.read_csv('lung_cancer_examples.csv')

# manipulation of smoking data to split inot positive and negative patients
yeslistsmokes = []
nolistsmokes = []
rowcount = 0
for outcome in dataset_1.iloc[:, -1]:
    if outcome == 1:
        yeslistsmokes.append(rowcount)
    else:
        nolistsmokes.append(rowcount)
    rowcount = rowcount + 1

positivecase_smoking = []
for yesvalue in yeslistsmokes:
    positivecase_smoking.append(dataset_1['Smokes'][yesvalue])
    
negcase_smoking = []
for novalue in nolistsmokes:
    negcase_smoking.append(dataset_1['Smokes'][novalue])
```

```python
# positive smoking only
plt.hist(positivecase_smoking, label = "positive cases")
plt.title('lung cancer smoking data')
plt.xlabel('Amount of smoking (larger = more)')
plt.ylabel('Frequency')
```

```python
# negative smoking only
plt.hist(negcase_smoking, label = "negative cases", facecolor = 'orange')
plt.xlabel('Amount of smoking (larger = more)')
plt.ylabel('Frequency')
plt.legend(loc="upper left")
```

```python
# over lap
plt.hist(positivecase_smoking, label = "positive cases")
plt.title('lung cancer smoking data')
plt.xlabel('Amount of smoking (larger = more)')
plt.ylabel('Frequency')

plt.hist(negcase_smoking, label = "negative cases")
plt.xlabel('Amount of smoking (larger = more)')
plt.ylabel('Frequency')
plt.legend(loc="upper left")
```

```python
# manipulation of alcohol data to split inot positive and negative patients

yeslistalc = []
nolistalc = []
rowcount = 0
for outcome in dataset_1.iloc[:, -1]:
    if outcome == 1:
        yeslistalc.append(rowcount)
    else:
        nolistalc.append(rowcount)
    rowcount = rowcount + 1

positivecase_alc = []
negcase_alc = []
for yesvalue in yeslistalc:
    positivecase_alc.append(dataset_1['Alkhol'][yesvalue])  
for novalue in nolistalc:
    negcase_alc.append(dataset_1['Alkhol'][novalue])
```

```python
# positive alcohol only
plt.hist(positivecase_alc, label = "positive cases", range = [0,8])
plt.title('lung cancer alcohol data')
plt.xlabel('Amount of alcohol (larger = more)')
plt.ylabel('Frequency')
```

```python
# negative alcohol only
plt.hist(negcase_alc, label = "negative cases", color = 'orange', range = [0,8])
plt.xlabel('Amount of alcohol (larger = more)')
plt.ylabel('Frequency')
plt.legend(loc = "upper left")
```

```python
# alcohol overlap
plt.hist(positivecase_alc, label = "positive cases")
plt.title('lung cancer alcohol data')
plt.xlabel('Amount of alcohol (larger = more)')
plt.ylabel('Frequency')

plt.hist(negcase_alc, label = "negative cases")
plt.xlabel('Amount of alcohol (larger = more)')
plt.ylabel('Frequency')
plt.legend(loc = "upper left")
```

```python
# getting means of patient data to use to compare user input to
dataset_2 = pd.read_csv('cancer patient data sets.csv')
dataset_2 = dataset_2.iloc[:, 4:21]
prediction_means = (dataset_2.describe().iloc[1:2, :])
prediction_means_list = prediction_means.values.tolist()
```

```python
# user input for risk %
print('Answer all questions on a scale of 1(no symptom) to 8(have symptom all of the time)')

patient_inputval_list = []
air_pollution = input('What level of air pollution exposure do you have?')
patient_inputval_list.append(int(air_pollution))
alc_use = input('What is your level of alcohol use?')
patient_inputval_list.append(int(alc_use))
dust_allergy = input('What is your level of dust allergy?')
patient_inputval_list.append(int(dust_allergy))
oc_hazard = input('What is your level of occupational hazards?')
patient_inputval_list.append(int(oc_hazard))
genetic_risk = input('What is your level of genetic risk?')
patient_inputval_list.append(int(genetic_risk))
chronic_lung_disease = input('What is your level of chronic lung disease?')
patient_inputval_list.append(int(chronic_lung_disease))
balanced_diet = input('What is your level of balanced diet (higher score = better)?')
patient_inputval_list.append(int(balanced_diet))
obesity = input('What is your level of obesity?')
patient_inputval_list.append(int(obesity))
smoking = input('What is your level of smoking?')
patient_inputval_list.append(int(smoking))
passive_smoker = input('What is your level of being a passive smoker?')
patient_inputval_list.append(int(passive_smoker))
chest_pain = input('What is your level of chest pain?')
patient_inputval_list.append(int(chest_pain))
coughing_blood = input('What is your level of coughing up blood?')
patient_inputval_list.append(int(coughing_blood))
fatigue = input('What is your level of fatigue?')
patient_inputval_list.append(int(fatigue))
weight_loss = input('What is your level of weight loss?')
patient_inputval_list.append(int(weight_loss))
short_breath = input('What is your level of shortness of breath?')
patient_inputval_list.append(int(short_breath))
wheezing = input('What is your level of wheezing?')
patient_inputval_list.append(int(wheezing))
swallowing_diff = input('What is your level of swallowing difficulty?')
patient_inputval_list.append(int(swallowing_diff))

print(patient_inputval_list)
print(prediction_means_list[0])
```

```python
# calculating risk %
risk_counter = 0
for value in range(17):
    if(patient_inputval_list[value] >= prediction_means_list[0][value] and value != 6):
        risk_counter = risk_counter + 1
    elif(patient_inputval_list[value] <= prediction_means_list[0][value] and value == 6):
        risk_counter = risk_counter + 1
print('your risk based on these questions is', risk_counter, '/17 or ', (risk_counter/17)*100, '%')
```
